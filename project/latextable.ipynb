```python
file = open('tags-markers-wx.txt', 'r').read()
```

```python
markers = file.split("\n\n")
```

```python
for marker in markers:
    mark = marker.split("\n")[0].split(', ')[0]
    tags = marker.split("\n")
    if len(tags) == 1:
        print('\hline')
        print(mark, '&', tags[0].split(', ')[1], '&', tags[0].split(', ')[2], '\\\\')
    elif len(tags) > 1:
        print('\hline')
        print('\multirow{', len(tags), '}{*}{', mark ,'}', '&', tags[0].split(', ')[1] ,'&', tags[0].split(', ')[2],'\\\\')
        for i in range(len(tags)-1):
            tokens = tags[i+1].split(', ')
            print('&', tokens[1] ,'&', tokens[2],'\\\\')
```
