```python
from collections import defaultdict, Counter
import nltk
```

```python
train_file = open('../assgn4/final_train.txt', 'r', encoding="utf-8")
traininput = train_file.read()
```

```python
def parse(input_):
    tags = []
    lexicons = []
    lemma = []
    pos= []
    sentences = input_.split("\n\n")
    for sentence in sentences:
        words = sentence.split("\n")
        for word in words:
            tokens = word.split("\t")
            try:
                tags.append(tokens[7])
                lexicons.append(tokens[1])
                pos.append(tokens[3])
                lemma.append(tokens[2])
            except:
                print(sentence)  
        lexicons.append("</s>")
        tags.append("</s>")
        lemma.append("</s>")
        pos.append("</s>")
    
    lexicons.pop()
    tags.pop()
    lemma.pop()
    pos.pop()
    return lexicons, tags, pos, lemma
```

```python
train_file = open('../assgn4/UD/hi_hdtb-ud-train.conllu.txt', 'r', encoding="utf-8")
traininput = train_file.read()
```

```python
def parse(input_):
    tags = []
    lexicons = []
    pos= []
    sentences = input_.split("\n\n")
    for sentence in sentences:
        words = sentence.split("\n")
        for word in words:
            try:
                if word[0] != '#':
                    tokens = word.split("\t")
                    try:
    #                     print(tokens[7])
                        tags.append(tokens[7])
                        lexicons.append(tokens[1])
                        pos.append(tokens[4])
                    except:
                        print("no", words)  
            except:
                print("wrong")
        lexicons.append("</s>")
        tags.append("</s>")
        lemma.append("</s>")
        pos.append("</s>")

        lexicons.pop()
        tags.pop()
        lemma.pop()
        pos.pop()
    return lexicons, tags, pos
```

```python
lexicons, tags, pos = parse(traininput)
```
