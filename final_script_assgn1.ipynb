```python
import stanfordnlp
nlp = stanfordnlp.Pipeline(lang='hi')
# Download hindi models: https://stanfordnlp.github.io/stanfordnlp/models.html#human-languages-supported-by-stanfordnlp
# on the model_path = ~/stanfordnlp_resources/
```

```python
def dependencies():
    corpus = open('corpus_20171055.txt', 'r')
    corpus = corpus.read()
    sentences = corpus.split('\n')
    count =0
    for sentence in sentences:
        count+=1
        print("<s>", count)
        doc = nlp(sentence)
        doc.sentences[0].print_dependencies()
    print(count)
```

```python
# run on terminal to get the output: python3 assgn1.py > dependency_asgn1.txt
## where assign1.py has **this cell and the above ones only**
dependencies()
```

```python
def preprocess():
    global token_id
    global dep_head
    global dep
    file = open('./siv/dep_1055.txt', 'r')
    file = file.read()
    sentences = file.split("<s>")
    for sentence in sentences:
        #print(type(sentence))
        lines = sentence.split("\n")
        for index, line in enumerate(lines):
            if len(line)> 0:
                line = line.replace('\')', "")
                line += "\', \'"
#                 print(index, line)
                labels = line.split("\', \'")
#                 print(labels[1])
                token_id.append(index)
                dep_head.append(labels[1])
                dep.append(labels[2])
    for i in range(0,1442):
        print(token_id[i], dep_head[i], dep[i])
```

```python
preprocess()
```

```python
def pos_preprocess():
    global token
    global pos
    global morph_coarse
    global morph_g
    global morph_n
    global morph_p
    global morph_do
    
    # tokenid, token, POS, morph
#     file = open('hindi.txt', 'r')
    file = open('./siv/h_20171055.txt', 'r')
    file = file.read()
    count=0 
    sentences = file.split("<s>")
    for sentence in sentences:
        lines = sentence.split("\n")
        for line in lines:
            labels = line.split("\t")
            if len(labels) == 9:
#                 print(count, labels)
                token.append(labels[0])
                pos.append(labels[2])
                morph_coarse.append(labels[4])
                morph_g.append(labels[5])
                morph_n.append(labels[6])
                morph_p.append(labels[7])
                morph_do.append(labels[8])
#     print(token_id)
    #print(count)
    for i in range(0,1442):
        print(token[i], pos[i], morph_coarse[i], morph_g[i], morph_n[i], morph_p[i], morph_do[i])
```

```python
pos_preprocess()
```
